```python

bigmart_sales.hist(figsize = (20,10))
```

```python
import matplotlib.pyplot as plt

x = bigmart_sales['Outlet_Establishment_Year']
y = bigmart_sales['Item_Weight']
plt.bar(x, y)
plt.xlabel('Outlet_Establishment_Year')
plt.ylabel('Item_Weight')
plt.title('Bar Chart Example')
plt.show() 
```

```python
x = bigmart_sales['Outlet_Establishment_Year']
y = bigmart_sales['Item_Outlet_Sales']
plt.bar(x, y)
plt.xlabel('Outlet_Establishment_Year')
plt.ylabel('Item_Outlet_Sales')
plt.title('Bar Chart Example')
plt.show() 
```

```python
grouped_data = bigmart_sales.groupby(["Outlet_Establishment_Year", "Item_Outlet_Sales"])
```

```python
mean_sales = grouped_data["Item_Outlet_Sales"].mean()
mean_sales
```

```python
x = bigmart_sales['Item_Type']
y = bigmart_sales["Item_Outlet_Sales"]
fig = plt.figure()
plt.bar(x, y)
plt.xlabel('Item_Type')
plt.ylabel('Item_Outlet_Sales')
plt.title('Bar Chart Example')
plt.xlabel("X-axis Label")
plt.xticks(rotation= 45 )
plt.subplots_adjust(bottom=0.2)
plt.show()
```
